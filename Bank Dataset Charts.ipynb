```python
import numpy as np
import matplotlib.pyplot as plt
import pandas
```

```python
adult_df = pandas.read_csv("AIF360 Debiasing Algo Analysis - Bank.csv")
adult_chart_df = pandas.read_csv("AIF360 Debiasing Algo Analysis - Bank Chart.csv")
```

```python
def create_graph(fairness_metric):
    colors = []
    for val in adult_df[fairness_metric]:
        if val > 0:
            colors.append('g')
        elif val < 0:
            colors.append('r')
        else:
            colors.append('b')

    plt.figure(figsize=(12,20))
    plt.title("Bank Dataset " + fairness_metric)
    plt.barh(adult_chart_df["Bank Dataset"], adult_chart_df[fairness_metric], 0.5, color=colors)
    plt.show()
```

```python
create_graph("false positive rate difference")
create_graph("false negative rate difference")
create_graph("error rate difference")
create_graph("false discovery rate difference")
create_graph("false omission rate difference")
create_graph("stat parity difference")
create_graph("priv base rate")
create_graph("unpriv base rate")
```
