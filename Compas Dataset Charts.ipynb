```python
import numpy as np
import matplotlib.pyplot as plt
import pandas
```

```python
compas_df = pandas.read_csv("AIF360 Debiasing Algo Analysis - Compas.csv")
compas_chart_df = pandas.read_csv("AIF360 Debiasing Algo Analysis - Compas Chart.csv")
```

```python
def create_value_graph(fairness_metric):
    colors = []
    for val in compas_df[fairness_metric]:
        if val > 0:
            colors.append('g')
        elif val < 0:
            colors.append('r')
        else:
            colors.append('b')

    plt.figure(figsize=(12,20))
    plt.title("Compas Dataset " + fairness_metric)
    plt.barh(compas_df["Compas"], compas_df[fairness_metric], 0.5, color=colors)
    plt.show()
```

```python
create_value_graph("false positive rate difference")
create_value_graph("false negative rate difference")
create_value_graph("error rate difference")
create_value_graph("false discovery rate difference")
create_value_graph("false omission rate difference")
create_value_graph("stat parity difference")
create_value_graph("priv base rate")
create_value_graph("unpriv base rate")
```
